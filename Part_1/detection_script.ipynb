```python
from ultralytics import YOLO
import json
import os

# Load trained model
# model = YOLO("runs/detect/train5/weights/best.pt")
model = YOLO(r"C:\Users\avira\Documents\Restart_Skills_v2025\GitHub\Technical_Assessment_Junior_ML_Engineer_Biz_Tech_Analytics\runs\detect\train5\weights\best.pt")


# Input/output folders
input_folder = r"C:\Users\avira\Documents\Restart_Skills_v2025\GitHub\Technical_Assessment_Junior_ML_Engineer_Biz_Tech_Analytics\Dataset\test"   # <-- use images folder
output_folder = "Part_1/Output"        # annotated images will be saved here
log_folder = "Part_1/Logs"             # logs in JSON

os.makedirs(output_folder, exist_ok=True)
os.makedirs(log_folder, exist_ok=True)

# Run inference and save annotated images
results = model.predict(
    source=input_folder,
    save=True,
    project=output_folder,
    name="annotated"
)

# Loop through results and save JSON logs
for r in results:
    detections = []
    img_name = os.path.basename(r.path)
    base_name, _ = os.path.splitext(img_name)
    log_file = os.path.join(log_folder, base_name + ".json")

    for box in r.boxes:
        cls_id = int(box.cls[0])
        label = model.names[cls_id]
        conf = float(box.conf[0])
        xyxy = [float(x) for x in box.xyxy[0].tolist()]

        detections.append({
            "label": label,
            "confidence": conf,
            "bbox": xyxy
        })

    log_data = {
        "filename": img_name,
        "detections": detections
    }

    with open(log_file, "w") as f:
        json.dump(log_data, f, indent=2)

print(f"✅ Detection complete. Annotated images in {output_folder}/annotated/, logs in {log_folder}/")
```
